EDA

Annual Salary based on Job Title

```python
Emp_JobTitle=df.groupby(['Annual Salary'], as_index= False)['Job Title'].sum().sort_values(by='Job Title')

sns.set(rc={'figure.figsize':(20,12)})
sns.barplot(x='Annual Salary',y='Job Title',data=Emp_JobTitle)
plt.xticks(rotation=90) 
plt.gca().invert_yaxis()
```

Gender based on Country

```python
sns.set(style='whitegrid')
sns.countplot(x='Gender',hue='Country',data=df)
plt.xlabel('Gender')
plt.ylabel('Country')
plt.show(close=True)
```

Exact Values

```python
ax=sns.countplot(x='Gender',hue='Country',data=df)
for bar in ax.containers:
    ax.bar_label(bar)
plt.xlabel('Gender')
plt.ylabel('Country')
plt.show(close=True)
```

```python
count_df = df.groupby(['Gender', 'Department']).size().reset_index(name='count')

sns.scatterplot(x='Gender',y='Department',hue='Department',data=count_df,size='count', sizes=(50, 500))
plt.xlabel('Gender')
plt.ylabel('Department')
plt.legend(loc='upper right',bbox_to_anchor=(1.336,1))
plt.show(close=True)
```
